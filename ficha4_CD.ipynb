```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

linguagem = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
pupularidade = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

ax.bar(linguagem, popularidade)
ax.set_ylabel("Pupularidade")
ax.set_xlabel("Linguagens")
ax.set_title("Popularidade das linguagens de programacao")
plt.minorticks_on()
plt.grid(True, which='major', color='red')
plt.grid(True, which='minor', color='green')
plt.show()

```
